##**Variables y Tipos de Datos**

* **¿Qué es una variable?**

* **Tipos de datos básicos**

### **Operaciones con variables**

    * Operadores aritméticos: `+`, `-`, `*`, `/`, `//` (división entera), `%` (módulo), `**` (potencia)
    * Operadores de comparación: `==` (igual), `!=` (no igual), `>`, `<`, `>=`, `<=`
    * Operadores lógicos: `and`, `or`, `not`
    * Operadores de asignación: `=`, `+=`, `-=`, `*=`, `/=`, `%=`, `**=`, `//=`, `&=`, `|=`, `^=`, `>>=`, `<<=`

####Ejemplos:

````python
# Operaciones Aritméticas
# Suma
num1 = 10
num2 = 5
suma = num1 + num2
print("Suma:", suma)

# Resta
resta = num1 - num2
print("Resta:", resta)

# Multiplicación
producto = num1 * num2
print("Producto:", producto)

# División
division = num1 / num2
print("División:", division)

# División Entera
division_entera = 11 // 2
print("División Entera:", division_entera)

# Módulo
modulo = 11 % 2
print("Módulo:", modulo)

# Potencia
potencia = num1 ** num2
print("Potencia:", potencia)

# Operadores de Comparación
a = 8
b = 12

igual = a == b
print("Igual:", igual)

no_igual = a != b
print("No Igual:", no_igual)

mayor_que = a > b
print("Mayor que:", mayor_que)

menor_que = a < b
print("Menor que:", menor_que)

mayor_o_igual = a >= 8
print("Mayor o igual:", mayor_o_igual)

menor_o_igual = b <= 12
print("Menor o igual:", menor_o_igual)

# Operadores Lógicos
x = True
y = False

# AND
and_op = x and y
print("AND:", and_op)

# OR
or_op = x or y
print("OR:", or_op)

# NOT
not_op_x = not x
print("NOT x:", not_op_x)
not_op_y = not y
print("NOT y:", not_op_y)

# Operadores de Asignación
# Asignación simple
numero = 10
print("Número:", numero)

# Suma y asignación
numero += 5
print("Número += 5:", numero)

# Resta y asignación
numero -= 3
print("Número -= 3:", numero)

# Multiplicación y asignación
numero *= 2
print("Número *= 2:", numero)

# División y asignación
numero /= 4
print("Número /= 4:", numero)

# Módulo y asignación
numero %= 5
print("Número %= 5:", numero)

# Potencia y asignación
numero **= 2
print("Número **= 2:", numero)
````

### **Conversión de tipos (Casting)**

    * Es necesaria para operar con variables de diferentes tipos.
    * Funciones de conversión:
        * `int()`: Convierte a entero.
        * `float()`: Convierte a punto flotante.
        * `str()`: Convierte a cadena.
        * `bool()`: Convierte a booleano.

####Ejemplos:

````python
#Conversion de tipos
num_str = "15"
num_int = int(num_str)  # num_int será 15 (entero)

print(f"\nel tipo de dato de num_str es {type(num_str)}")
print(f"el tipo de dato de num_int es {type(num_int)}")

pi = 3.14159
pi_entero = int(pi)    # pi_entero será 3 (entero)

print(f"\nel tipo de dato de pi es {type(pi)}")
print(f"el tipo de dato de pi_entero es {type(pi_entero)}")

edad = 25
edad_str = str(edad)  # edad_str será "25" (cadena)

print(f"\nel tipo de dato de edad es {type(edad)}")
print(f"el tipo de dato de edad_str es {type(edad_str)}")

valor = 0
valor_bool = bool(valor) # valor_bool será False

print(f"\nel tipo de dato de valor es {type(valor)}")
print(f"el tipo de dato de valor_bool es {type(valor_bool)}")
````

##**Funciones Predefinidas Comunes**

* **Funciones de entrada/salida**

    * `print()`: Imprime en la consola.
    * `input()`: Lee la entrada del usuario.
* **Funciones matemáticas**

    * `abs()`: Valor absoluto.
    * `pow()`: Potencia.
    * `round()`: Redondea un número.
    * Funciones del módulo `math`: `sqrt()` (raíz cuadrada), `ceil()` (redondea hacia arriba), `floor()` (redondea hacia abajo), `pi`, `e`, etc.
* **Funciones de cadenas**

    * `len()`: Longitud de una cadena.
    * `str.upper()`: Convierte a mayúsculas.
    * `str.lower()`: Convierte a minúsculas.
    * `str.strip()`: Elimina espacios en blanco al inicio y al final.
    * `str.find()`: Busca una subcadena.
    * `str.replace()`: Reemplaza una subcadena.

    * `str.join()`: Une una lista de cadenas en una sola cadena.
* **Otras funciones útiles**

    * `type()`: Devuelve el tipo de dato de una variable.

###Ejemplos

````python
# Funciones de entrada/salida
# print(): Imprime en la consola
print("Hola, esto es un mensaje.")

# input(): Lee la entrada del usuario
nombre = input("Ingresa tu nombre: ")
print("Hola,", nombre)

# Funciones matemáticas
# abs(): Valor absoluto
numero = -7
valor_absoluto = abs(numero)
print("Valor absoluto de", numero, "es", valor_absoluto)

# pow(): Potencia
base = 2
exponente = 3
potencia = pow(base, exponente)
print(base, "elevado a la", exponente, "es", potencia)

# round(): Redondea un número
numero_decimal = 3.14159
redondeo_2_decimales = round(numero_decimal, 2)
print(numero_decimal, "redondeado a 2 decimales es", redondeo_2_decimales)

# Funciones del módulo math
import math

raiz_cuadrada = math.sqrt(25)
print("Raíz cuadrada de 25:", raiz_cuadrada)

pi = math.pi
print("Valor de pi:", pi)

# Funciones de cadenas
# len(): Longitud de una cadena
cadena = "Python"
longitud = len(cadena)
print("Longitud de la cadena", cadena, "es", longitud)

# str.upper(): Convierte a mayúsculas
cadena_minusculas = "hola"
cadena_mayusculas = cadena_minusculas.upper()
print(cadena_minusculas, "en mayúsculas es", cadena_mayusculas)

# str.lower(): Convierte a minúsculas
cadena_mayusculas = "MUNDO"
cadena_minusculas = cadena_mayusculas.lower()
print(cadena_mayusculas, "en minúsculas es", cadena_minusculas)

# str.strip(): Elimina espacios en blanco al inicio y al final
cadena_con_espacios = "  Hola  "
cadena_sin_espacios = cadena_con_espacios.strip()
print("Cadena con espacios:", cadena_con_espacios)
print("Cadena sin espacios:", cadena_sin_espacios)

# str.find(): Busca una subcadena
texto = "Esta es una cadena de ejemplo"
posicion = texto.find("cadena")
print("Posición de 'cadena':", posicion)

# str.replace(): Reemplaza una subcadena
texto_original = "Manzana es una fruta"
texto_reemplazado = texto_original.replace("Manzana", "Banana")
print("Texto original:", texto_original)
print("Texto reemplazado:", texto_reemplazado)

# str.split(): Divide una cadena en una lista de subcadenas
cadena_palabras = "uno,dos,tres"
lista_palabras = cadena_palabras.split(",")
print("Lista de palabras:", lista_palabras)

# str.join(): Une una lista de cadenas en una sola cadena
lista_nombres = ["Ana", "Juan", "Maria"]
cadena_nombres = ", ".join(lista_nombres)
print("Cadena de nombres:", cadena_nombres)

# Otras funciones útiles
# type(): Devuelve el tipo de dato de una variable
variable1 = 10
variable2 = "Hola"
print("Tipo de variable1:", type(variable1))
print("Tipo de variable2:", type(variable2))

# id(): Devuelve la "identidad" (dirección de memoria) de un objeto
objeto1 = [1, 2, 3]
objeto2 = objeto1
print("ID de objeto1:", id(objeto1))
print("ID de objeto2:", id(objeto2))  # Mismo ID porque es la misma lista

# help(): Invoca el sistema de ayuda integrado (útil en modo interactivo)
# help(print) # Muestra la ayuda de la función print
````

##**Estructuras de Datos**
**Listas**

* **Definición:** Una lista es una colección **ordenada** y **mutable** de elementos.
* "Ordenada" significa que los elementos tienen una posición definida y se mantienen en ese orden.
* "Mutable" significa que puedes modificar la lista después de su creación (agregar, eliminar, cambiar elementos).
* Las listas pueden contener elementos de diferentes tipos de datos (números, cadenas, otras listas, etc.).

**Tuplas**

* **Definición:** Una tupla es una colección **ordenada** e **inmutable** de elementos.
* "Ordenada" tiene el mismo significado que en las listas: los elementos tienen una secuencia.
* "Inmutable" es la clave: una vez que se crea una tupla, no puedes cambiar sus elementos (no puedes agregar, eliminar ni modificar).
* Las tuplas también pueden contener elementos de diferentes tipos.

**Sets (Conjuntos)**

* **Definición:** Un set es una colección **no ordenada** de elementos **únicos**.
* "No ordenada" significa que los elementos no tienen una posición fija; el orden puede variar.
* "Únicos" es fundamental: un set no puede contener elementos duplicados; si intentas agregar uno, se ignora.
* Los sets son útiles para operaciones matemáticas de conjuntos (unión, intersección, etc.).

**Diccionarios**

* **Definición:** Un diccionario es una colección **no ordenada** de pares **clave-valor**.
* En lugar de acceder a los elementos por su posición (como en listas o tuplas), accedes a ellos por una "clave".
* Cada clave debe ser única dentro del diccionario.
* Los diccionarios son mutables: puedes agregar, eliminar o modificar pares clave-valor.
* Son muy eficientes para buscar valores rápidamente dado su clave.

###**Listas**

* Creación: `[ ]` o `list()`
* Acceso a elementos: por índice (empezando desde 0).
* Slicing: `[inicio:fin:paso]`
* Métodos: `append()`, `insert()`, `remove()`, `pop()`, `index()`, `count()`, `sort()`, `reverse()`, `copy()`, `clear()`
* Ejemplos:

```python
    frutas = ["manzana", "banana", "cereza"]
    frutas.append("naranja")
    print(frutas[1])
    print(frutas[1:3])
  ```

````python
#   Métodos de Listas

print("--- Listas ---")
frutas = ["manzana", "banana", "cereza"]

#   append(elemento): Agrega un elemento al final de la lista.
frutas.append("naranja")
print("Después de append:", frutas)  #   Output: ['manzana', 'banana', 'cereza', 'naranja']

#   insert(índice, elemento): Inserta un elemento en una posición específica.
frutas.insert(1, "kiwi")
print("Después de insert:", frutas)  #   Output: ['manzana', 'kiwi', 'banana', 'cereza', 'naranja']

#   remove(elemento): Elimina la primera aparición del elemento.
frutas.remove("banana")
print("Después de remove:", frutas)  #   Output: ['manzana', 'kiwi', 'cereza', 'naranja']

#   pop(índice): Elimina y devuelve el elemento en el índice dado. Si no se proporciona el índice, elimina y devuelve el último elemento.
elemento_eliminado = frutas.pop(2)
print("Después de pop:", frutas)  #   Output: ['manzana', 'kiwi', 'naranja']
print("Elemento eliminado:", elemento_eliminado)  #   Output: cereza

#   index(elemento): Devuelve el índice de la primera aparición del elemento.
indice_manzana = frutas.index("manzana")
print("Índice de manzana:", indice_manzana)  #   Output: 0

#   count(elemento): Devuelve el número de veces que aparece el elemento en la lista.
frutas.append("manzana")
cantidad_manzanas = frutas.count("manzana")
print("Cantidad de manzanas:", cantidad_manzanas)  #   Output: 2

#   sort(): Ordena la lista in situ (modifica la lista original).
frutas.sort()
print("Después de sort:", frutas)  #   Output: ['kiwi', 'manzana', 'manzana', 'naranja']

#   reverse(): Invierte la lista in situ.
frutas.reverse()
print("Después de reverse:", frutas)  #   Output: ['naranja', 'manzana', 'manzana', 'kiwi']

#   copy(): Devuelve una copia superficial de la lista.
frutas_copia = frutas.copy()
frutas_copia.append("sandía")
print("Copia:", frutas_copia)  #   Output: ['naranja', 'manzana', 'manzana', 'kiwi', 'sandía']
print("Original:", frutas)  #   Output: ['naranja', 'manzana', 'manzana', 'kiwi']  (No se modifica)

#   clear(): Elimina todos los elementos de la lista.
frutas.clear()
print("Después de clear:", frutas)  #   Output: []

print("\n")
````

###**Tuplas**

* Creación: `( )` o `tuple()`
* Acceso a elementos: por índice.
* Slicing: igual que las listas.
* Métodos: `count()`, `index()`
* Ejemplos:

```python
    colores = ("rojo", "verde", "azul")
    print(colores[0])
```

````python
#   Métodos de Tuplas

print("--- Tuplas ---")
colores = ("rojo", "verde", "azul", "rojo")

#   count(elemento): Devuelve el número de veces que aparece el elemento en la tupla.
cantidad_rojo = colores.count("rojo")
print("Cantidad de 'rojo':", cantidad_rojo)  #   Output: 2

#   index(elemento): Devuelve el índice de la primera aparición del elemento.
indice_verde = colores.index("verde")
print("Índice de 'verde':", indice_verde)  #   Output: 1

print("\n")
````

###**Sets (Conjuntos)**

* Creación: `{ }` o `set()`
* Operaciones de conjuntos: unión (`|`), intersección (`&`), diferencia (`-`), diferencia simétrica (`^`)
* Métodos: `add()`, `remove()`, `discard()`, `pop()`, `clear()`, `update()`
* Ejemplos:

```python
    numeros = {1, 2, 3, 3, 4, 5}  # Los duplicados se eliminan
    pares = {2, 4, 6}
    print(numeros | pares)  # Unión
    print(numeros & pares)  # Intersección
```


````python
#   Métodos de Sets

print("--- Sets ---")
numeros = {1, 2, 3, 4, 5}
pares = {2, 4, 6, 8}

#   add(elemento): Agrega un elemento al set.
numeros.add(6)
print("Después de add:", numeros)  #   Output: {1, 2, 3, 4, 5, 6}  (Orden no garantizado)

#   remove(elemento): Elimina el elemento del set. Si el elemento no existe, lanza un error (KeyError).
numeros.remove(5)
print("Después de remove:", numeros)  #   Output: {1, 2, 3, 4, 6}

#   discard(elemento): Elimina el elemento del set si está presente. No lanza un error si el elemento no existe.
numeros.discard(7)  #   No pasa nada porque 7 no está
numeros.discard(4)
print("Después de discard:", numeros)  #   Output: {1, 2, 3, 6}

#   pop(): Elimina y devuelve un elemento arbitrario del set. Lanza un error si el set está vacío.
elemento_arbitrario = numeros.pop()
print("Después de pop:", numeros)  #   Output: Por ejemplo: {2, 3, 6} (Elemento eliminado no predecible)
print("Elemento pop:", elemento_arbitrario)

#   clear(): Elimina todos los elementos del set.
numeros.clear()
print("Después de clear:", numeros)  #   Output: set()

#   update(iterable): Agrega elementos de otro iterable (lista, tupla, set, etc.) al set.
numeros.update([5, 6, 7])
print("Después de update:", numeros)  #   Output: {5, 6, 7}

#   Operaciones de conjuntos (devuelven nuevos sets)
print("Unión:", numeros | pares)  #   {2, 4, 5, 6, 7, 8}
print("Intersección:", numeros & pares)  #   {4, 6}
print("Diferencia (numeros - pares):", numeros - pares)  #   {5, 7}
print("Diferencia (pares - numeros):", pares - numeros)  #   {2, 8}
print("Diferencia simétrica:", numeros ^ pares)  #   {2, 5, 7, 8}

print("\n")
````

###**Diccionarios**

* Definición: Colección no ordenada de pares clave-valor.
* Creación: `{ clave: valor }` o `dict()`
* Acceso a valores: por clave.
* Métodos: `get()`, `keys()`, `values()`, `items()`, `update()`, `pop()`, `popitem()`, `clear()`, `copy()`, `setdefault()`
* Ejemplos:

```python
  persona = {"nombre": "Ana", "edad": 28, "ciudad": "México"}
  print(persona["nombre"])
  persona["edad"] = 29
  persona["ocupacion"] = "Ingeniera"
```

````python
#   Métodos de Diccionarios

print("--- Diccionarios ---")
persona = {"nombre": "Ana", "edad": 30, "ciudad": "Madrid"}

#   get(clave, valor_por_defecto): Devuelve el valor asociado a la clave. Si la clave no existe, devuelve el valor_por_defecto (o None si no se proporciona).
nombre = persona.get("nombre")
print("Nombre:", nombre)  #   Output: Ana
ocupacion = persona.get("ocupacion", "Desconocido")
print("Ocupación:", ocupacion)  #   Output: Desconocido

#   keys(): Devuelve una vista de las claves del diccionario.
claves = persona.keys()
print("Claves:", claves)  #   Output: dict_keys(['nombre', 'edad', 'ciudad'])

#   values(): Devuelve una vista de los valores del diccionario.
valores = persona.values()
print("Valores:", valores)  #   Output: dict_values(['Ana', 30, 'Madrid'])

#   items(): Devuelve una vista de los pares (clave, valor) del diccionario.
items = persona.items()
print("Items:", items)  #   Output: dict_items([('nombre', 'Ana'), ('edad', 30), ('ciudad', 'Madrid')])

#   update(otro_diccionario): Actualiza el diccionario con los pares clave-valor de otro diccionario o iterable de pares clave-valor.
persona.update({"edad": 31, "ocupacion": "Ingeniera"})
print("Después de update:", persona)  #   Output: {'nombre': 'Ana', 'edad': 31, 'ciudad': 'Madrid', 'ocupacion': 'Ingeniera'}

#   pop(clave, valor_por_defecto): Elimina y devuelve el valor asociado a la clave. Si la clave no existe, devuelve el valor_por_defecto (o lanza un error si no se proporciona).
ciudad = persona.pop("ciudad")
print("Después de pop:", persona)  #   Output: {'nombre': 'Ana', 'edad': 31, 'ocupacion': 'Ingeniera'}
print("Ciudad:", ciudad)  #   Output: Madrid

#   popitem(): Elimina y devuelve un par (clave, valor) arbitrario del diccionario. Lanza un error si el diccionario está vacío.
par_eliminado = persona.popitem()
print("Después de popitem:", persona)  #   Por ejemplo: {'nombre': 'Ana', 'edad': 31} (Par eliminado no predecible)
print("Par eliminado:", par_eliminado)

#   clear(): Elimina todos los elementos del diccionario.
persona.clear()
print("Después de clear:", persona)  #   Output: {}

#   copy(): Devuelve una copia superficial del diccionario.
persona_original = {"nombre": "Ana", "edad": 30}
persona_copia = persona_original.copy()
persona_copia["edad"] = 31
print("Original:", persona_original)  #   Output: {'nombre': 'Ana', 'edad': 30}
print("Copia:", persona_copia)  #   Output: {'nombre': 'Ana', 'edad': 31}

#   setdefault(clave, valor_por_defecto): Devuelve el valor asociado a la clave. Si la clave no existe, inserta la clave con el valor_por_defecto y devuelve el valor_por_defecto (o None si no se proporciona).
persona = {"nombre": "Ana", "edad": 30}
ocupacion = persona.setdefault("ocupacion", "Desconocido")
print("Después de setdefault:", persona)  #   Output: {'nombre': 'Ana', 'edad': 30, 'ocupacion': 'Desconocido'}
print("Ocupación:", ocupacion)  #   Output: Desconocido
````
