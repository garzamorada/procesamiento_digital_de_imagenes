```python
from skimage import io, filters, measure, feature
from matplotlib import pyplot as plt
```

```python
! wget 'https://www2.eecs.berkeley.edu/Research/Projects/CS/vision/bsds/BSDS300/html/images/plain/normal/color/239096.jpg'
```

```python
img = io.imread('239096.jpg')
io.imshow(img)
```

```python
# Convert the image to grayscale
gray_img = io.imread('239096.jpg', as_gray=True)
io.imshow(gray_img)
```

```python
edges = feature.canny(gray_img)
io.imshow(edges)

```
