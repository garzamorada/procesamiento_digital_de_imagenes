##**Importación de módulos:**
* Importa los módulos necesarios de `skimage` (io, filters, measure, feature) y `matplotlib.pyplot`.

```python
from skimage import io, filters, measure, feature
from matplotlib import pyplot as plt
```

##**Carga de imagen:**
* Descarga una imagen de ejemplo usando `wget`.
* Carga la imagen con `io.imread()` y la muestra con `io.imshow()`.

```python
! wget 'https://www2.eecs.berkeley.edu/Research/Projects/CS/vision/bsds/BSDS300/html/images/plain/normal/color/239096.jpg'
```

```python
img = io.imread('239096.jpg')
io.imshow(img)
```

##**Filtros:**
* Aplica varios filtros de `skimage.filters` (por ejemplo, `sobel`, `gaussian`) a la imagen y muestra los resultados.
* **Comentario:** Los filtros se utilizan para realzar características de la imagen, como bordes o suavizado.

```python
filtered_img = filters.gaussian(img, sigma=3)
io.imshow(filtered_img)
```

##**Conversion a escala de grises**
* Convierte de RGB a escala de grises

```python
# Convert the image to grayscale
gray_img = io.imread('239096.jpg', as_gray=True)
io.imshow(gray_img)
```

##**Mediciones:**
* Realiza mediciones en la imagen (por ejemplo busqueda de contorno).
* **Comentario:** `measure` permite extraer propiedades cuantitativas de las regiones de la imagen.

```python
# Busca contornos de distintas areas
contours = measure.find_contours(gray_img, 0.2)

# Mostar los resultados
fig, ax = plt.subplots()
ax.imshow(gray_img, cmap=plt.cm.gray)

for contour in contours:
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)  # Plotting contours

ax.axis('image')  # Set aspect ratio to match image
ax.set_xticks([])  # Remove x-axis ticks
ax.set_yticks([])  # Remove y-axis ticks
plt.show()
```

##**Características:**
* Extrae características de la imagen (por ejemplo, bordes con `feature.canny`).
* **Comentario:** `feature` proporciona herramientas para la extracción de características relevantes para el análisis de imágenes.

```python
edges = feature.canny(gray_img)
io.imshow(edges)

```
